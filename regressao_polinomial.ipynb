## 游늳 Avalia칞칚o e An치lise dos Modelos

### Regress칚o Linear

```python
# Defini칞칚o do Holdout de 50%
df1 = df.copy()

Y = 'Pre칞o'
X = ['Tamanho_m2','Num_Banheiros','Renda_Comprador','Dist_Centro','Renda_Vari치vel','Num_Quartos','Idade_Imovel','Num_Filhos', 'Temp_Anterior', 'Anuncios_Imoveis']

X_train, X_test, Y_train, Y_test = train_test_split(df1[X], df1[Y], test_size=0.5, random_state=42)
X_train.head()
```

```python
# Calculando o R dos sample de treino e de teste
modelo_linear = LinearRegression()
modelo_linear.fit(X_train, Y_train)

print('Regress칚o Linear')
print(f'R Score do sample de treino: {modelo_linear.score(X_train, Y_train)}')
print(f'R Score do sample de teste: {modelo_linear.score(X_test, Y_test)}')
```

```python
# Sumariza칞칚o e resultados da regress칚o de treino
X_train = sm.add_constant(X_train)
modelo_linear = sm.OLS(Y_train, X_train).fit()
print(modelo_linear.summary())
```

Todas as vari치veis apresentam p-valor baixo (< 0.05), o que indica que as vari치veis s칚o significativas!

```python
# Sumariza칞칚o e resultados da regress칚o de treino
X_test = sm.add_constant(X_test)
modelo_linear = sm.OLS(Y_test, X_test).fit()
print(modelo_linear.summary())
```

Todas as vari치veis apresentam p-valor baixo (< 0.05), o que indica que as vari치veis s칚o significativas!

### Regress칚o Polinomial

```python
# Defini칞칚o do Holdout de 50%
df2 = df.copy()

Y = 'Pre칞o'
X = ['Tamanho_m2','Num_Banheiros','Renda_Comprador','Dist_Centro','Renda_Vari치vel','Num_Quartos','Idade_Imovel','Num_Filhos', 'Temp_Anterior', 'Anuncios_Imoveis']

poly_f = PolynomialFeatures(degree = 10)
X_poly = poly_f.fit_transform(df2[X].to_numpy())

X_train, X_test, Y_train, Y_test = train_test_split(X_poly, df2[Y], test_size=0.5, random_state=42)
```

```python
# Modelagem da Regress칚o Polinomial
modelo_polinomial = LinearRegression()

modelo_polinomial.fit(X_train, Y_train)
A = modelo_polinomial.score(X_train, Y_train)
print(A)


modelo_polinomial.fit(X_test, Y_test)
B = modelo_polinomial.score(X_test, Y_test)
print(B)
```

```python
# Verifica칞칚o da diferen칞a absoluta do r2 de treino e r2 de teste
abs(A - B)
```

Verificamos que a diferen칞a absoluna entre r2 de treino e r2 de teste 칠 menor do que 0.2

A an치lise do R, aliada ao controle da diferen칞a entre treino e teste, mostrou que um modelo polinomial com grau adequado pode capturar rela칞칫es n칚o lineares sem perder capacidade de generaliza칞칚o.
