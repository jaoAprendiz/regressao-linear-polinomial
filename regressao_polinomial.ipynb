## üìà Avalia√ß√£o e An√°lise dos Modelos

### Regress√£o Linear

```python
# Defini√ß√£o do Holdout de 50%
df1 = df.copy()

Y = 'Pre√ßo'
X = ['Tamanho_m2','Num_Banheiros','Renda_Comprador','Dist_Centro','Renda_Vari√°vel','Num_Quartos','Idade_Imovel','Num_Filhos', 'Temp_Anterior', 'Anuncios_Imoveis']

X_train, X_test, Y_train, Y_test = train_test_split(df1[X], df1[Y], test_size=0.5, random_state=42)
X_train.head()
```

```python
# Calculando o R¬≤ dos sample de treino e de teste
modelo_linear = LinearRegression()
modelo_linear.fit(X_train, Y_train)

print('Regress√£o Linear')
print(f'R¬≤ Score do sample de treino: {modelo_linear.score(X_train, Y_train)}')
print(f'R¬≤ Score do sample de teste: {modelo_linear.score(X_test, Y_test)}')
```

```python
# Sumariza√ß√£o e resultados da regress√£o de treino
X_train = sm.add_constant(X_train)
modelo_linear = sm.OLS(Y_train, X_train).fit()
print(modelo_linear.summary())
```

Todas as vari√°veis apresentam p-valor baixo (< 0.05), o que indica que as vari√°veis s√£o significativas!

```python
# Sumariza√ß√£o e resultados da regress√£o de treino
X_test = sm.add_constant(X_test)
modelo_linear = sm.OLS(Y_test, X_test).fit()
print(modelo_linear.summary())
```

Todas as vari√°veis apresentam p-valor baixo (< 0.05), o que indica que as vari√°veis s√£o significativas!

### Regress√£o Polinomial

```python
# Defini√ß√£o do Holdout de 50%
df2 = df.copy()

Y = 'Pre√ßo'
X = ['Tamanho_m2','Num_Banheiros','Renda_Comprador','Dist_Centro','Renda_Vari√°vel','Num_Quartos','Idade_Imovel','Num_Filhos', 'Temp_Anterior', 'Anuncios_Imoveis']

poly_f = PolynomialFeatures(degree = 10)
X_poly = poly_f.fit_transform(df2[X].to_numpy())

X_train, X_test, Y_train, Y_test = train_test_split(X_poly, df2[Y], test_size=0.5, random_state=42)
```

```python
# Modelagem da Regress√£o Polinomial
modelo_polinomial = LinearRegression()

modelo_polinomial.fit(X_train, Y_train)
A = modelo_polinomial.score(X_train, Y_train)
print(A)


modelo_polinomial.fit(X_test, Y_test)
B = modelo_polinomial.score(X_test, Y_test)
print(B)
```

```python
# Verifica√ß√£o da diferen√ßa absoluta do r2 de treino e r2 de teste
abs(A - B)
```

Verificamos que a diferen√ßa absoluna entre r2 de treino e r2 de teste √© menor do que 0.2

## Conclus√£o

Com base nos experimentos realizados, a Regress√£o Linear apresentou resultados consistentes, com vari√°veis estatisticamente significativas que explicam parte da variabilidade do pre√ßo. J√° a Regress√£o Polinomial demonstrou maior capacidade de ajuste, mas exige cautela para evitar overfitting, especialmente em graus muito elevados. A an√°lise do R¬≤, aliada ao controle da diferen√ßa entre treino e teste, mostrou que um modelo polinomial com grau adequado pode capturar rela√ß√µes n√£o lineares sem perder capacidade de generaliza√ß√£o. Em resumo, a escolha do modelo ideal depende do equil√≠brio entre simplicidade, interpretabilidade e capacidade de previs√£o.
